```python
import json

def merge_histories(resp1, resp2, t_threshold=25):
    h1 = resp1["history"]
    h2 = resp2["history"]

    merged = []

    # iterate over same block count (same index)
    for item1, item2 in zip(h1, h2):
        d1, t1, p1 = item1["datetime"], item1["t"], item1["p"]
        t2, p2 = item2["t"], item2["p"]
        if abs(t1 - t2) <= t_threshold:
            merged.append({
                "datetime": d1,
                "total_p": p1 + p2
            })

    return merged

# resp1 = json.loads(response.text)
# resp2 = json.loads(response2.text)

result = merge_histories(response, response2)

print(result)
```

```python
# Putting everything tgt (event slug --> market + clobTokenId + history)
import requests
import json
from datetime import datetime, timezone, timedelta

UTC8 = timezone(timedelta(hours=8))

slug = "btc-updown-15m-1768221900"
url_w_id = f"https://gamma-api.polymarket.com/events/slug/{slug}"

event = requests.get(url_w_id).json()

# print(json.dumps(event, indent=2, ensure_ascii=False))
print(f"Event: {event['id']}, {event['title']}")

markets = event['markets']
print(f"Markets in this event: {len(markets)} with id {[m['id'] for m in markets]}")

for i, market in enumerate(markets):
    market_id = market['id']
    url_w_id = f"https://gamma-api.polymarket.com/markets/{market_id}"

    market = requests.get(url_w_id).json()

    clobTokenIds = json.loads(market['clobTokenIds']) # returns as str, so convert it back to json
    print(f"clobTokenIds in market {market_id}: {clobTokenIds}")
    
    assert len(clobTokenIds) == 2

    clobTokenId = clobTokenIds[0]
    interval = "max"
    fidelity = "1"
    url = f"https://clob.polymarket.com/prices-history?market={clobTokenId}&interval={interval}&fidelity={fidelity}"

    response = requests.get(url)
    response.raise_for_status()
    response = response.json()
    for item in response["history"]:
        item["datetime"] = datetime.fromtimestamp(
            item["t"], tz=UTC8
        ).strftime("%Y-%m-%d %H:%M:%S")
    print("History for yes:", response)

    clobTokenId2 = clobTokenIds[1]
    url2 = f"https://clob.polymarket.com/prices-history?market={clobTokenId2}&interval={interval}&fidelity={fidelity}"

    response2 = requests.get(url2)
    response2.raise_for_status()
    response2 = response2.json()
    for item in response2["history"]:
        item["datetime"] = datetime.fromtimestamp(
            item["t"], tz=UTC8
        ).strftime("%Y-%m-%d %H:%M:%S")
    print("History for no:", response2)
    
```

```python
# Script to extract all history events
import requests
import time
import json
import os

def get_existing_total(output_path):
    if not os.path.exists(output_path):
        return 0
    with open(output_path, "r", encoding="utf-8") as f:
        return sum(1 for _ in f)

def get_all_events(output_path="events.jsonl", max_retries=5):
    base_url = "https://gamma-api.polymarket.com/events"
    limit = 100

    # Resume logic
    total = get_existing_total(output_path)
    offset = total
    print(f"Resuming from offset={offset}")

    retries = 0

    # Append mode (important!)
    with open(output_path, "a", encoding="utf-8") as f:
        while True:
            try:
                start1 = time.time()

                response = requests.get(
                    base_url,
                    params={"limit": limit, "offset": offset},
                    timeout=10
                )
                response.raise_for_status()
                data = response.json()

                # No more data
                if not data:
                    print("No more events.")
                    break

                for event in data:
                    f.write(json.dumps(event, ensure_ascii=False) + "\n")
                    total += 1

                f.flush()

                offset += limit
                retries = 0  # reset retry counter on success

                end1 = time.time()
                print(
                    f"Time: {round(end1 - start1, 2)}s | "
                    f"Total events: {total} | Offset: {offset}"
                )

            except Exception as e:
                retries += 1
                print(
                    f"Error at offset={offset} "
                    f"(retry {retries}/{max_retries}): {e}"
                )

                if retries >= max_retries:
                    print(
                        f"Giving up after {max_retries} retries at offset={offset}. "
                        f"Progress safely saved."
                    )
                    break

                time.sleep(2)  # small backoff before retry

    return total


if __name__ == "__main__":
    start = time.time()
    total_events = get_all_events()
    end = time.time()
    print(
        f"Done. Total time: {round(end - start, 2)}s | "
        f"Total events saved: {total_events}"
    )
```
