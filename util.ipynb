```python
# Standardize results.csv
import csv

# Input CSV file path
file_path = './data/results.csv'

# Open the file and read the data
with open(file_path, mode='r', newline='') as infile:
    reader = csv.reader(infile)
    rows = list(reader)

# Modify the "UP" or "Down" values in place
for row in rows:
    if row[1].strip().lower() == 'up':
        row[1] = 'U'
    elif row[1].strip().lower() == 'down':
        row[1] = 'D'

# Write the modified data back to the same file
with open(file_path, mode='w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(rows)

print(f"File has been modified successfully at {file_path}")
```

```python
# Convert between UTC and datetime
import time
from datetime import datetime

# Get current UTC timestamp
utc_timestamp = time.time()

# Convert to readable format
utc_datetime = datetime.utcfromtimestamp(utc_timestamp)

print(f"UTC timestamp in seconds: {utc_timestamp}")
print(f"UTC time in readable format: {utc_datetime}")
```
